#### Handling missing values

```python
df_titanic.isnull()         # will list a table (row * column) where null values are present
```

```python
df_titanic.isnull().any()       # will list out columns where null values are present
```

```python
df_titanic_renamed = df_titanic.rename(columns = {
    'Survived': 'How many survived'
})
df_titanic_renamed
```

```python
df1 = pd.DataFrame({
    'Key': ['Alpha', 'Beta', 'Charlie'],
    'Value': [1, 2, 3],
})

print(df1)

df2 = pd.DataFrame({
    'Key': ['Alpha', 'Beta', 'Delta'],
    'Value': [4, 5, 6],
})

print(df2)

pd.merge(df1, df2, on = 'Key', how = 'inner')           # merges only common rows i.e, A & B
```

```python
pd.merge(df1, df2, on = 'Key', how = 'outer')       # there are also joins like 'left' and 'right' like SQL Joins
```

```python
# writes & creates data to a csv file 
df_titanic_renamed.to_csv('df_titanic_renamed.csv')
```

```python
# creating a pickle (binary file cannot be read) used for serialization from an existing Dataframe object

df_titanic_pkl = df_titanic.to_pickle('dataframe.pkl')
```

```python
# deserializing back to pandas object
pd.read_pickle('dataframe.pkl')
```
